```python
import numpy as np
import pandas as pd
battle = pd.read_csv('battles.csv')
```

```python
battle.head()
```

```python
battle.shape
```

```python
battle.rename(columns={'attacker-1':'primary_attacker'},inplace=True)
battle.head()
```

```python
battle.rename(columns={'defender-1':'primary_defender'},inplace=True)
battle.head()
```

```python
battle['attacker_king'].value_counts()
```

```python
battle['location'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='attacker_king',y='attacker_size',data=battle)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='defender_king',y='defender_size',data=battle)
plt.show()
```

```python
sns.countplot(x=battle['attacker_king'],hue=battle['battle_type'])
plt.show()
```
